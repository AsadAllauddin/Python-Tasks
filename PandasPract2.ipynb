```python
df.dropna(axis=0,inplace=True)
print(df)
```

```python
#Dropping the duplicates in the whole dataset
df.drop_duplicates(inplace=True)
df
```

```python
#Show the null values via description column
df[df['Description'].isnull() == True]
#Apparently all the null and duplicate values have dropped 
```

```python
#Also, no negative value in "Qualtity" and "price" cloumn
df[(df['Quantity'] <=0) & (df['Price']<0.0)]
```

```python

df[df['Price']==0]
```

```python
#how many rows we have where the price are = 0 
df[df['Price']==0].shape
```

```python
#dropping the rows where the prices are 0
df.drop(df[df['Price']==0].index, axis=0, inplace=True)
df
```

```python
#Checking if we have negative value in our "Quantity" cloumn
df[df['Quantity']<=0]
#And apparently we have 8872 rows with negative value
```

```python
#Using data visualization to show "Customer ID" and "Quantity" in scatter graph
df.plot(x='Customer ID',y='Quantity', kind='scatter')
plt.show()
```

```python
#Sorting values according to "Quantity" and we found positive and negative outliners
df.sort_values(by='Quantity', inplace=True)
df
```

```python
# removing the outliers from dataset 
df.drop(df.iloc[:3].index, axis =0 , inplace=True)
df.drop(df.iloc[-2:].index, axis =0 , inplace=True)
df
```

```python
df.plot(x='Customer ID',y='Quantity', kind='scatter')
plt.show()
#the scatter graph changes it alignments
```

```python
df.plot(x='Customer ID',y='Price', kind='scatter')
plt.show()
```

```python
lrgst=df.nlargest(1,'Price')
lrgst
```

```python
# removing the highest price outlier
df.drop(labels=lrgst.index, inplace=True)
```

```python
#graph after dropping largest-price outliner
df.plot(x='Customer ID',y='Price', kind='scatter')
plt.show()
```

```python
#Adding a new column named "Total Price"
df['Total Price'] = df['Quantity']* df['Price']
df
```

```python
countrywiseprice= df.groupby('Country')['Total Price'].sum().sort_values(ascending = False)[:5]
countrywiseprice
```

```python
g= df.groupby('Country')['Total Price'].sum().sort_values(ascending = False)[:5].plot(kind='bar')
g
```

```python
max_index=df['Total Price'].idxmax()
max_index
```

```python
df['Month'] =  df['InvoiceDate'].dt.month
df.head()
```

```python
#Sales by month
df.groupby('Month')['Total Price'].sum().sort_values(ascending = False).plot(kind='bar', color= 'orange')
```

```python
#Take a look at the top 5 customers
df.groupby('Customer ID')['Total Price'].sum().sort_values(ascending = False).iloc[:5].plot(kind='bar' ,color='green')
```

```python
#top 25 products 
df.groupby('Description')['Quantity'].sum().sort_values(ascending = False).iloc[:25].plot(kind='bar')
```

```python
df['Hour'] = df['InvoiceDate'].dt.hour
df.head()
```

```python
hours = [hour for hour, d in df.groupby('Hour')]
plt.plot(hours , df.groupby('Hour').count())
plt.xticks(hours)
plt.grid()
plt.show()
#top busy hours and it seems it's 11am to 2pm a/c to this data set
```
